# Create the Category Data, Save it to a Dataframe, and Export to a CSV FILE

```python
# pandas dependency
import pandas as pd
from pathlib import Path
import numpy as np
import csv
```

```python
# Create the Category DataFrame:

crowdfunding_data_df = pd.read_excel("Resources/crowdfunding.xlsx")

category_data_df = crowdfunding_data_df[['category & sub-category']].copy()

category_data_df = category_data_df.rename(columns={"category & sub-category": "cat_subcat"})

category_data_df.cat_subcat.unique()

unique_cat_subcat = category_data_df.cat_subcat.unique()

#category_cat_dict = {}

```

```python
cat_subcat_df = pd.DataFrame(columns=['Category', 'Sub_Category'])
#cat_subcat_df = pd.DataFrame(columns=['A', 'B', 'C', 'D', 'E'])

cat_subcat_df.shape
for row in unique_cat_subcat:
    row_list = row.split("/")
    cat_subcat_df.loc[len(cat_subcat_df)] = row_list

#print(cat_subcat_df)
```

```python
#category_only_df = cat_subcat_df[['Category']].copy()

#    cat_subcat_df.loc[len(cat_subcat_df)] = row_list
```

```python
#  Create unique CAT IDs for each unique Category
category_only_df = cat_subcat_df[['Category']].copy()
unique_cats = category_only_df.Category.unique()
catnum = 0
cat_catid_df = pd.DataFrame(columns=['category_id', 'category'])
for cat in unique_cats:
    catnum = catnum + 1
    catid = 'cat' +  str(catnum)
    addcat = []
    addcat.append(catid)
    addcat.append(cat)
#    print(addcat)
    cat_catid_df.loc[len(cat_catid_df)] = addcat
#    print(cat, catid)
    
```

```python
cat_catid_df
```

```python
#  Create unique SUB-CAT IDs for each unique Sub-Category
subcategory_only_df = cat_subcat_df[['Sub_Category']].copy()
unique_subcats = subcategory_only_df.Sub_Category.unique()
subcatnum = 0
subcat_scatid_df = pd.DataFrame(columns=['subcategory_id', 'subcategory'])
for subcat in unique_subcats:
    subcatnum = subcatnum + 1
    subcatid = 'subcat' + str(subcatnum)
    saddcat = []
    saddcat.append(subcatid)
    saddcat.append(subcat)
#    print(saddcat)
    subcat_scatid_df.loc[len(subcat_scatid_df)] = saddcat
```

```python
subcat_scatid_df
```

```python
# Export the data to CSV Files

#This file contains the CATEGORY DAtA
cat_catid_df.to_csv('Resources/crowdfunding_Category_Data.csv',index=False)

#This file contains the SUB-CATEGORY DATA
subcat_scatid_df.to_csv('Resources/crowdfunding_SUB_Category_Data.csv',index=False)
```
